# Libreria numpy

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Colonne -> 0
# Righe -> 1
# Globale -> senza parametri

mat = np.array([[1,-5,0],
                [4,3,6],
                [5,8,-7],
                [10,6,-12]])

print(f"Massimo per righe: {mat.max(0)}\n") # massimo per colonne
print(f"Minimo per righe: {mat.min(1)}\n") # minimo per righe
print(f"Somma globale: {mat.sum()}\n") # somma di tutti gli elementi
print(f"Indici dei massimi per righe: {mat.argmax(1)}") # indici dei massimi per righe
```

# Libreria Matplotlib

- Definire un vettore arbitrario di valori estratti dal dominio della funzione;
- Calcolare i rispettivi punti utilizzando la forma analitica riportata sopra;
- Eseguire il plot dei valori mediante matplotlib;

```python
a = -100
b = 100

x = np.linspace(a, b, 300) # Campioniamo in maniera lineare 300 punti tra -100 e 100
y = pow(x, 2) # Calcoliamo il quadrato di ognuno dei punti di x
```

### Mostrare il grafico

```python
plt.plot(x, y) # stampa le coordinate x e y nel piano cartesiano
plt.xlabel("ascisse") #imposta una etichetta per l'asse x
plt.ylabel("ordinate") #imposta una etichetta per l'asse y
plt.title("f(x) = xÂ²") #imposta il titolo del grafico
plt.show() #mostra il grafico
```

### Controllo dei tick e della griglia

```python
plt.plot(x,y) # stampa le coordinate x e y nel piano cartesiano

plt.xlim([-50, 50]) # visualizziamo i valori x solo tra -50 e 50
plt.ylim([0, 2500]) # e i valori y solo tra 0 e 2500

plt.xticks([-40,0,40]) # inseriamo solo -40, 0 e 40 come "tick" sull'asse x
plt.yticks([1000,2000]) # solo 1000 e 2000 come tick y

plt.grid() # aggiungiamo una griglia
plt.show()
```
